```
import json
```

```
class Sentiment:
        positive = 'POSITIVE'
        neutral = 'NEUTRAL'
        negative = 'NEGATIVE'
```

```
class Review:
    def __init__(self, text, score):
        self.text = text
        self.score = score
        self.sentiment = self.get_sentiment()

    def get_sentiment(self):
        if self.score <=2:
            return Sentiment.negative
        elif self.score == 3:
            return Sentiment.neutral
        else:
            return Sentiment.positive
```

```
filehandle = open('Books_small.json')
print(filehandle.readline())

all_reviews = list()
for line in filehandle:
    review = json.loads(line)
    all_reviews.append(Review(review['reviewText'], review['overall']))

print(all_reviews[0].sentiment)
print(len(all_reviews))
```

### Train Test Split

```
from sklearn.model_selection import train_test_split
train, test = train_test_split(all_reviews, test_size = 0.33, random_state = 42)
print(len(train), len(test))
```

```
x_train = [x.text for x in train]
y_train = [x.sentiment for x in train]
print(x_train)
print(y_train)

x_test = [x.text for x in test]
y_test = [x.sentiment for x in test]
```

### Bags of words vectorization

```
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

# below line is combination of vectorizer.fit(x_train) and x_train_vectors = vectorizer.transform(x_train)
x_train_vectors = vectorizer.fit_transform(x_train)
x_test_vectors = vectorizer.transform(x_test)

print('text= ', x_train[0])

#prints columns with values>=1 for row 0
print(x_train_vectors[0])
```

## Classification

#### Linear SVM

```
from sklearn.svm import SVC

linear_svm_model = SVC(kernel = 'linear')
linear_svm_model.fit(x_train_vectors, y_train)
```

```
print(x_test[19])
linear_svm_model.predict(x_test_vectors[19])
#predicts neutral despite of negative
```

#### Decision Trees

```
from sklearn.tree import DecisionTreeClassifier
dtree_model = DecisionTreeClassifier()
dtree_model.fit(x_train_vectors, y_train)
```

```
print(x_test[19])
dtree_model.predict(x_test_vectors[19])
```

#### Naive Bayes

```
from sklearn.naive_bayes import GaussianNB
naive_bayes_model = GaussianNB()
naive_bayes_model.fit(x_train_vectors.toarray(), y_train)
```

```
print(x_test[19])
naive_bayes_model.predict(x_test_vectors[19].toarray())
#o/p should be negative
```

#### Logistic Regression

```
from sklearn.linear_model import LogisticRegression
logistic_regression_model = LogisticRegression(max_iter=200)
logistic_regression_model.fit(x_train_vectors, y_train)
```

```
print(x_test[19])
logistic_regression_model.predict(x_test_vectors[19])
```

### Compare models

```
print(f'Linear SVM Model- {linear_svm_model.score(x_test_vectors, y_test)}')
print(f'Decision Tree Model- {dtree_model.score(x_test_vectors, y_test)}')
print(f'Naive Bayes Model- {naive_bayes_model.score(x_test_vectors.toarray(), y_test)}')
print(f'Logistic Regression Model- {logistic_regression_model.score(x_test_vectors, y_test)}')
```
